### Cleaning & Formatting

```python
df = df.dropna(subset=['RecipeInstructions', 'RecipeIngredientParts', 'Name'])
```

```python
text_columns = ['Name', 'AuthorName', 'DatePublished', 'Description', 'RecipeCategory', 'Keywords', 'RecipeIngredientParts', 'RecipeInstructions']
for col in text_columns:
    df[col] = df[col].astype(str).str.strip()
```

```python
numeric_columns = ['Calories', 'FatContent', 'CarbohydrateContent', 'ProteinContent', 'FiberContent', 'SugarContent', 'AggregatedRating', 'ReviewCount']
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df = df.dropna(subset=['Calories', 'FatContent', 'CarbohydrateContent', 'ProteinContent', 'FiberContent', 'SugarContent'])
```

```python
df['is_rated'] = ~df['AggregatedRating'].isna()
```

```python
df['RecipeServings'] = df['RecipeServings'].fillna(0).astype(int)
```

```python
def clean_list_string(s):
    s = s.replace('c(', '').replace(')', '').replace('"', '')
    return [item.strip() for item in s.split(',') if item.strip()]

def format_instructions(instruction_list):
    return " ".join(f"{i+1}. {step}" for i, step in enumerate(instruction_list))

def parse_duration(duration_str):
    if pd.isna(duration_str) or not isinstance(duration_str, str) or not duration_str.startswith('PT'):
        return "N/A", 0
    
    try:
        duration_str = duration_str.replace('PT', '')
        hours, minutes = 0, 0
        
        if 'H' in duration_str:
            hours_part = duration_str.split('H')[0]
            hours = int(hours_part) if hours_part else 0
            duration_str = duration_str.split('H')[1] if 'H' in duration_str else ''
        
        if 'M' in duration_str:
            minutes_part = duration_str.split('M')[0]
            minutes = int(minutes_part) if minutes_part else 0
        
        total_minutes = hours * 60 + minutes
        
        if hours and minutes:
            display = f"{hours} hours {minutes} minutes"
        elif hours:
            display = f"{hours} hours"
        elif minutes:
            display = f"{minutes} minutes"
        else:
            display = "N/A"
        
        return display, total_minutes
    except (ValueError, IndexError) as e:
        print(f"Error parsing duration: {duration_str}, Error: {e}")
        return "N/A", 0
```

```python
df['Keywords'] = df['Keywords'].apply(clean_list_string)
df['Ingredients'] = df['RecipeIngredientParts'].apply(clean_list_string)
df['Instructions'] = df['RecipeInstructions'].apply(clean_list_string)
df['Instructions'] = df['Instructions'].apply(format_instructions)
df['RecipeCategory'] = df['RecipeCategory'].str.lower()
```

```python
df['PrepTimeDisplay'], df['PrepTimeMinutes'] = zip(*df['PrepTime'].apply(parse_duration))
df['CookTimeDisplay'], df['CookTimeMinutes'] = zip(*df['CookTime'].apply(parse_duration))
df['TotalTimeDisplay'], df['TotalTimeMinutes'] = zip(*df['TotalTime'].apply(parse_duration))
```

```python
ordered_filtered_cols = [
    # identity / Metadata
    'RecipeId', 'Name', 'AuthorId', 'AuthorName', 'DatePublished', 'RecipeCategory', 'Keywords', 'AggregatedRating', 'ReviewCount', 'is_rated',

    # main Content
    'Description', 'RecipeIngredientParts', 'RecipeInstructions', 'Ingredients', 'Instructions',

    # nutrition
    'Calories', 'FatContent', 'CarbohydrateContent', 'ProteinContent', 'FiberContent', 'SugarContent', 

    # Optional / Derived
    'PrepTimeDisplay', 'PrepTimeMinutes', 'CookTimeDisplay', 'CookTimeMinutes', 'TotalTimeDisplay', 'TotalTimeMinutes', 'RecipeServings'
]

df = df[ordered_filtered_cols]
```

### Create Docuemnts

```python
# fucntion to combine fields into a single document per recipe
def create_document(row):
    rating_text = (
        f"{row['AggregatedRating']} (from {int(row['ReviewCount'])} reviews)"
        if not pd.isna(row['AggregatedRating'])
        else "Not yet rated"
    )

    time_text = f"Total Time: {row['TotalTimeDisplay']}"
    servings_text = (
        str(int(row['RecipeServings']))
        if row['RecipeServings'] > 0
        else "Not specified"
    )
    return (
        f"Recipe: {row['Name']}\n"
        f"Category: {row['RecipeCategory']}\n"
        f"Rating: {rating_text}\n"
        f"Keywords: {', '.join(row['Keywords'])}\n"
        f"Description: {row['Description']}\n"
        f"Ingredients: {', '.join(row['Ingredients'])}\n"
        f"Instructions: {row['Instructions']}\n"
        f"Nutrition (per serving): Calories: {row['Calories']}, "
        f"Fat: {row['FatContent']}g, Carbs: {row['CarbohydrateContent']}g, "
        f"Protein: {row['ProteinContent']}g, Fiber: {row['FiberContent']}g, "
        f"Sugar: {row['SugarContent']}g\n"
        f"{time_text}, Serves: {servings_text}"
    )

# apply to create a list of documents
documents = df.apply(create_document, axis=1).tolist()
```
